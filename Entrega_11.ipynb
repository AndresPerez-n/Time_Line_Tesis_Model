```python
import numpy as np
```

```python
import pandas as pd
```

```python
class ACP:
  def __init__(self,acp, scale= True):
    self.acp = acp
    self.scale = scale
    self.eigvecs = None
  def inercia(self):
    if self.scale == True:
      acp_scal = np.zeros(acp.shape)
      for i in range(acp_scal.shape[1]):
        acp_scal[:,i] = (acp[:,i] - acp[:,i].mean())/acp[:,i].std()
      self.acp = acp_scal 
    else: 
      self.acp = self.acp
    cov = self.acp.T.dot(self.acp)/(self.acp.shape[0])
    eigvals, self.eigvecs = np.linalg.eig(cov)
    var_exp = {i:i/sum(eigvals) for i in eigvals}
    m_inercia = pd.DataFrame({'eigvals':eigvals,
                               'eigvals_acum': np.cumsum(eigvals),
                               'var_exp': var_exp.values(),
                               'v_expl_acum': np.cumsum(list(var_exp.values()))})
    
    return m_inercia.T

  def projec(self):
    Names = ['0']*acp.shape[1]
    for i in range(1,acp.shape[1] + 1):
      Names[i-1] = str(i)
    project = self.acp.dot(self.eigvecs[:,:self.acp.shape[1]])
    project = pd.DataFrame(project, columns = Names)
    return project
```

```python
from sklearn.datasets import load_breast_cancer
acp = load_breast_cancer()["data"]
```

```python
ACP= ACP(acp)
```

```python
ACP.inercia ()
```

```python
ACP.projec()
```
